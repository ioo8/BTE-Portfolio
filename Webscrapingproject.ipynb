```python

import yfinance as yf
import pandas as pd

# Extract Tesla's stock data from yfinance
tesla = yf.Ticker("TSLA")
tesla_stock_data = tesla.history(period = '1mo')
print(tesla_stock_data.head())

tesla_stock_data.to_csv('tesla_stock_data.csv')

```

```python
# Extract Tesla's revenue data from webscraping
import requests
from bs4 import BeautifulSoup

url = 'https://stockanalysis.com/stocks/tsla/revenue/'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
revenue_data = soup.find_all('table')[0]
tesla_revenue_df = pd.read_html(str(revenue_data))[0]

print("Tesla Revenue Data:")
print(tesla_revenue_df.head())

# Save as CSV
tesla_revenue_df.to_csv('tesla_revenue_data.csv')
```

```python
# Extract Gamestops's stock data from yfinance
gamestop = yf.Ticker("GME")
gamestop_stock_data = gamestop.history(period = '1mo')
print(gamestop_stock_data.head())

gamestop_stock_data.to_csv('gamestop_stock_data.csv')
```

```python
# Extract Gamestops's revenue data from webscraping
import requests
from bs4 import BeautifulSoup

url = 'https://stockanalysis.com/stocks/gme/revenue/'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
revenue_data = soup.find_all('table')[0]
gamestop_revenue_df = pd.read_html(str(revenue_data))[0]

print("Gamestop History Data:")
print(gamestop_revenue_df.head())

# Save as CSV
gamestop_revenue_df.to_csv('gamestop_revenue_data.csv')
```

```python
import matplotlib.pyplot as plt

# Creating Dashboards
tesla_stock_data = pd.read_csv('tesla_stock_data.csv', index_col = 0)
tesla_revenue_data = pd.read_csv('tesla_revenue_data.csv')
gamestop_stock_data = pd.read_csv('gamestop_stock_data.csv', index_col = 0)
gamestop_revenue_data = pd.read_csv('gamestop_revenue_data.csv')

# Tesla Stock Data
plt.figure(figsize = (14, 7))
plt.plot(tesla_stock_data.index, tesla_stock_data['Close'], label = 'Tesla Stock Price')
plt.title("Tesla Stock Data")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()

plt.savefig('Tesla_stock_dashboard.png')

# Tesla Revenue Data
plt.figure(figsize = (14, 7))
plt.plot(tesla_revenue_data['Fiscal Year End'], tesla_revenue_data['Revenue'], label = 'Tesla Revenue')
plt.title("Tesla Revenue Data")
plt.xlabel('Fiscal Year End')
plt.ylabel('Revenue')
plt.legend()
plt.grid()

plt.savefig('Tesla_revenue_dashboard.png')

# GME Stock Data
plt.figure(figsize = (14, 7))
plt.plot(gamestop_stock_data.index, gamestop_stock_data['Close'], label = 'Gamestop Stock Price')
plt.title("Gamestop Stock Data")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()

plt.savefig('gamestop_stock_dashboard.png')

#GME Revenue Data

plt.figure(figsize = (14, 7))
plt.plot(gamestop_revenue_data['Fiscal Year End'], gamestop_revenue_data['Revenue'], label = 'Gamestop Revenue')
plt.title("Gamestop Revenue Data")
plt.xlabel('Fiscal Year End')
plt.ylabel('Revenue')
plt.legend()
plt.grid()

plt.savefig('Gamestop_revenue_dashboard.png')
```
